```python
# Chọn các biến định tính cần thống kê
qualitative_vars = ['AQI Category', 'CO AQI Category', 'Ozone AQI Category', 'NO2 AQI Category', 'PM2.5 AQI Category']

# Tính tần số xuất hiện của các giá trị định tính
qualitative_counts = df[qualitative_vars].apply(pd.Series.value_counts)

print("Qualitative Statistics:")
print(qualitative_counts)
```

```python
#phân tích đơn biến
# Chọn biến cần phân tích (AQI, CO, Ozone, NO2, PM2.5)
variables = ['AQI Value', 'CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']

# Lặp qua từng biến để thực hiện phân tích đơn biến
for variable in variables:
    # Xóa các giá trị thiếu trong biến
    variable_data = df[variable].dropna()

    # Tạo biểu đồ histogram
    plt.figure(figsize=(8, 6))
    sns.histplot(variable_data, kde=True, color='blue')
    plt.xlabel(variable)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {variable}')
    plt.show()

    # Tạo biểu đồ boxplot
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=variable_data, color='orange')
    plt.xlabel(variable)
    plt.title(f'Boxplot of {variable}')
    plt.show()

    # Tính các thống kê mô tả cơ bản
    stats = variable_data.describe()
    print(f"Descriptive Statistics for {variable}:\n{stats}\n")
```

```python
#phân tích nhị biến
col = ['AQI Value', 'CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']
fig = plt.figure(figsize=(15, 5))

for i in range(len(col)):
    plt.subplot(1, 5, i+1)
    plt.title(col[i])
    sns.histplot(data=df, x=col[i])
plt.tight_layout()
plt.show()
```

```python
#nhị biến, biểu đồ hình nến tìm hiểu?
fig = plt.figure(figsize=(15,10))

for i in range(len(col)):
    plt.subplot(2,3,i+1)
    plt.title(col[i])
    sns.boxplot(data=df,x=df[col[i]],y=df['AQI Category'])
plt.tight_layout()
plt.show()
```

```python
#phân tích nhị biến
# Group by country and calculate the maximum AQI value for each country
top_countries = df.groupby('Country')['AQI Value'].max().nlargest(15)

# Tạo biểu đồ cột nằm ngang
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis", orient='h')
plt.xlabel('Maximum AQI Value')
plt.ylabel('Country')
plt.title('Top 15 Countries with Highest Maximum AQI Values')
plt.tight_layout()
plt.show()
```

```python
#phân tích nhị biến
# Group by country and calculate the maximum CO AQI value for each country
top_countries_co = df.groupby('Country')['CO AQI Value'].max().nlargest(15)

# Tạo biểu đồ cột cho chỉ số không khí CO
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries_co.values, y=top_countries_co.index, palette="viridis")
plt.xlabel('Maximum CO AQI Value')
plt.ylabel('Country')
plt.title('Top 15 Countries with Highest Maximum CO Values')
plt.tight_layout()
plt.show()
```
